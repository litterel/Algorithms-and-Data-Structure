```python
def bubble_sort(a):
    n = len(a)
    if n <= 1:
        return
        # return a
    else:
        step = 1
        while step < n:
            for i in range(n-step):
                if a[i] > a[i+1]:
                    temp = a[i]
                    a[i] = a[i+1]
                    a[i+1] = temp
            step += 1
        # return a


def select_sort(a):
    n = len(a)
    if n <= 1:
        pass
    else:
        for i in range(0, n-1):
            min = i
            for j in range(i+1, n):
                if a[j] < a[min]:
                    min = j
            temp = a[i]
            a[i] = a[min]
            a[min] = temp


def insert_sort(a):
    n = len(a)
    if n <= 1:
        return

    for i in range(1, n):
        now = a[i]
        position = 0
        for j in range(i-1, -1, -1):
            if a[j] > now:
                a[j+1] = a[j]
            else:
                position = j+1
                break
        a[position] = now


def shell_sort(a):
    n = len(a)
    if n <= 1:
        return
    
    d = int(sqrt(n))
    while d >= 1:
        for i in range(d):
            subarray = a[i:n:d]
            insert_sort(subarray)
        d = int(d/2)


def merge_sort(a, low, high):
    if low == high-1:
        if a[low] > a[high]:
            temp = a[low]
            a[low] = a[high]
            a[high] = temp
    elif low >= high:
        pass
    else:
        mid = int((low+high)/2)
        merge_sort(a, low, mid)
        merge_sort(a, mid+1, high)


def quick_sort(a, low, high):

    if low>= high or low > len(a)-1:
        return
    ax = a[low]
    i = low
    j = high
    while(i < j):
        while a[j] >= ax and i < j:
            j -= 1

        while a[i] <= ax and i < j:
            i += 1
        temp = a[i]
        a[i] = a[j]
        a[j] = temp

    a[low] = a[i]
    a[i] = ax

    quick_sort(a, low, i-1)
    quick_sort(a, i+1, high)


def qsort_nonrecur(a, low, high):
    index_stacks = []
    index_stacks.append((low, high))
    while len(index_stacks) != 0:
        left, right = index_stacks.pop()
        
        if left >= right or left > len(a)-1:
            pass
        else:
            ax = a[left]
            i = left
            j = right
            while(i < j):
                while a[j] >= ax and i < j:
                    j -= 1

                while a[i] <= ax and i < j:
                    i += 1
                temp = a[i]
                a[i] = a[j]
                a[j] = temp
                
            a[left] = a[i]
            a[i] = ax

            index_stacks.append((left, i-1))
            index_stacks.append((i+1, high))


def mergesort(a, low, high):
    pass
```
